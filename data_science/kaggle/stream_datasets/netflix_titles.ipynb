```python
# by 2022 these people are 14 yo, so
ratings2008 = ["G", "PG", "PG-13", "TV-Y", "TV-Y7", "TV-Y7 FV", "TV-G", "TV-PG"]
df2008 = df.loc[(df["rating"].isin(ratings2008))]
df2008
```

```python
# # additionnally let's drop unnecessary column(s)
df2008.drop(columns=["show_id"], inplace=True)
```

Okay, so far so good. But maybe let's narrow this down a bit further.<br />
Considering at this age range children are most likely to enjoy animated and/or comedic productions more often, whereas most pre-teenagers tend to like comic-books and/or fantasy ones. So let's work that out.

```python
# let's take a look at the type ("listed_in") columns
types = df2008["listed_in"].values.tolist()
types = set(types)
print(len(types))
```

```python
# narrow it down (I'm sure this can and should be done in a more efficient manner)
types_narrdown = set()
for item in types:
    new_items = item.split()
    for new_item in new_items:
        if new_item != "&" and new_item[-1] not in [","]:
            types_narrdown.add(new_item)
print(types_narrdown)
```

```python
# arbitrarily select types that would be the preferred / most suited ones
# and list all prod in those categories
# (since df2008 is already rating filtered out there's no need to include any conditions about it)
types2008 = "adventure|comedy|children|kids|anime|teen|fantasy|sci-fi"
df2008 = df2008.loc[df2008["listed_in"].str.contains(types2008, flags=re.I, regex=True)]
df2008.reset_index(drop=True, inplace=True)
```

Now, maybe let's talk about materials length. Movies with 100+ mins running time are less likely to captivate younger people. Same for shows with more than 4 seasons. So we gonna work that out.

```python
# let's list all movies under 100 mins and TV shows with 4 seasons or less
## but first let's rearrange the data so this column is of integer type
df2008["duration"].dtype
```

```python
# well, since the column contains data with both numerical and string values, let's get rid of the strings
## using the .split method in a lambda
df2008["duration"] = df2008["duration"].apply(lambda x: x.split()[0])
# then transform to integer
df2008["duration"] = pd.to_numeric(df2008["duration"], errors="coerce").fillna(0)
```

```python
df2008
```

```python
# now let's filter out
df2008 = df2008.loc[((df2008["type"] == "Movie") & (df2008["duration"] <= 100)) | ((df2008["type"] == "TV Show") & (df2008["duration"] <= 4))]
df2008
```
